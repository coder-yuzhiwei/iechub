```python
import faspell
```

```python
spell_checker = faspell.SpellChecker()
```

```python
epoch = 4
```

```python
# 第一版Ocr-Chinese
with open('check_data/test.txt','r') as f_o, open('check_data/truth.txt','r') as f_t:
    sentences = []
    truth = f_t.readlines()
    
    reader = csv.reader(f_o, delimiter='\t')
    for sent in reader:
        sentences.append(sent)
    sentences.sort()
    sentences = [sent[1] for sent in sentences]
    faspell.repeat_non_test(sentences, spell_checker, epoch)
    f = open(f'history.json', 'r', encoding='utf-8')
    result = json.load(f)
    print('纠正结果')
    for i in range(len(truth)):
        print('**********senence: {}**********'.format(i))
        print('OCR:\t{}'.format(result[i]['original_sentence']))
        print('纠错:\t{}'.format(result[i]['correction_history'][epoch-1]))
        print('实际:\t{}'.format(truth[i]))
```

```python
epoch = 3
```

```python
# 第二版OCR-Polar-Pass
with open('/Users/jas/Desktop/OCR-Polar-Pass/list.txt','r') as f_o, open('/Users/jas/Desktop/Seal/seal_data.txt','r') as f_t:
    truth = []
    for sent in csv.reader(f_t):
        truth.append(sent)
    
    origin = []
    csv_reader = csv.reader(f_o, delimiter='\t')
    for sent in csv_reader:
        num = re.findall(r'\d+', sent[0])[0]
        i = '0'*(3-len(num)) + num
        origin.append([i, sent[1]])
    origin.sort()
    sentences = [i[1] for i in origin]
    faspell.repeat_non_test(sentences, spell_checker, epoch)
    f = open(f'history.json', 'r', encoding='utf-8')
    result = json.load(f)
    print('纠正结果')
    right = 0
    wrong = 0
    img_index = [i[0] for i in origin]
    for i, index in enumerate(img_index):
        index = int(index)
        print('**********seal: {}**********'.format(index))
        print('OCR:\t{}'.format(result[i]['original_sentence']))
        print('纠错:\t{}'.format(result[i]['correction_history'][epoch-1]))
        print('实际:\t{}'.format(truth[index-1][0]))
        correct = list(result[i]['correction_history'][epoch-1])
        groud_truth = list(truth[i][0])
        for a,b in zip(correct, groud_truth):
            if a==b:
                right += 1
            else:
                wrong += 1
    print('正确率:\t{}'.format(right/(right+wrong)))
```
